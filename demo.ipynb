# Comparación RAG vs GraphRAG

```python
pregunta = "¿Cuáles son los principales puntos que hacen los políticos jóvenes vs los políticos mayores respecto los precios de las viviendas?"
df_recuperado = recuperar_top_k(
    pregunta_usuario=pregunta,
    modelo_e5=modelo_e5
)
respuesta = generar_respuesta_rag(pregunta, df_recuperado, cliente)
display(Markdown(respuesta))
```

```python
tema_a_buscar = "precios de las viviendas"
df_recuperado_menos50 = contexto_filtro_edad(tema_a_buscar, 50, "<", modelo_e5)
df_recuperado_mas50 = contexto_filtro_edad(tema_a_buscar, 50, ">=", modelo_e5)

df_recuperado = pd.concat([df_recuperado_menos50.iloc[:5], df_recuperado_mas50.iloc[:5]])
df_recuperado = df_recuperado.sort_values(by="score_similitud").reset_index(drop=True)
df_recuperado
```

```python
pregunta_llm = "¿Cuáles son los principales puntos que hacen los políticos jóvenes vs los políticos mayores respecto los precios de las viviendas?"
respuesta = generar_respuesta_rag(pregunta_llm, df_recuperado, cliente)
display(Markdown(respuesta))
```

Solo usar RAG provoca que el llm tenga inferir la edad de los políticos en base al contexto, lo que hace que ser jóven quede a criterio del llm. Un ejemplo de esto es el senador Juan Castro que el llm lo cataloga de político joven, siendo que en la actualidad tiene 66 años dificilmente pudiendo catalogarse como jóven al lado de políticos de menos de 40. Por lo mismo, no queda claro si la visión entregada es realmente la de políticos jóvenes o no. 

En cambio, el uso de GraphRAG le permitió al llm discernir las posturas entre grupos etarios claramente, gracias a que ahora está a dispocision la edad al momento de la intervención. Las respuestas dadas entregan una visión clara de ambos grupos etarios sin la incertidumbre del caso anterior.
